```python
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier

import matplotlib.pyplot as plt

import warnings 

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('customer_segmentation.csv')

print(data.head())

```

```python
plt.scatter(data['annual_income_k'],data['segment'])
plt.show()
plt.scatter(data['spending_score'],data['segment'])
plt.show()
plt.scatter(data['age'],data['segment'])
plt.show()
```

```python
features = data[['age','annual_income_k','spending_score']]

result = data['segment']
```

```python
model = KNeighborsClassifier(n_neighbors=5)

model.fit(features,result)
```

```python
cus_age = float(input("Enter Customer Age : "))
cus_annual_salary = float(input("Enter Customer Annual Customer Salary : "))
cus_spending = float(input("Enter Customer Spending Score : "))

predicted_data = [[cus_age,cus_annual_salary,cus_spending]]

prediction_customer_type = model.predict(predicted_data)

if prediction_customer_type == 0:
    print("The Customer Budget is Low it is in category 0")
elif prediction_customer_type == 1:
        print("The Customer Budget is Medium it is in category 1")
elif prediction_customer_type == 2:
    print("Customer Budget is High it is in category 2")
```
